```python
from mmdet.datasets import build_dataset
import numpy as np
import matplotlib.pyplot as plt
import shutil
import os
from tqdm.notebook import tqdm
```

Run this notebook to create the VOC-OS and COCO-OS datasets mentioned in the paper. VOC-OS has the first 15 classes known, last 5 unknown. COCO-OS has the first 50 classes known, last 30 unknown.

# Pascal VOC

```python
VOC_CLASSES = (  # always index 0
    'aeroplane', 'bicycle', 'bird', 'boat',
    'bottle', 'bus', 'car', 'cat', 'chair',
    'cow', 'diningtable', 'dog', 'horse',
    'motorbike', 'person', 'pottedplant',
    'sheep', 'sofa', 'train', 'tvmonitor')
```

```python
#setup the voc2007 and 2012 training datasets

test_pipeline = [
    dict(type='LoadImageFromFile'),
    dict(
        type='MultiScaleFlipAug',
        img_scale=(608, 608),
        flip=False,
        transforms=[
            dict(type='ImageToTensor', keys=['img']),
            dict(type='Collect', keys=['img'])
        ])
]

vocData2007 = dict(samples_per_gpu = 4, workers_per_gpu = 4,
               train = dict(type = 'VOCDataset',
                ann_file=  'data/VOCdevkit/VOC2007/ImageSets/Main/trainval.txt',
                img_prefix= 'data/VOCdevkit/VOC2007/',
                pipeline=test_pipeline))

vocData2012 = dict(samples_per_gpu = 4, workers_per_gpu = 4,
               train = dict(type = 'VOCDataset',
                ann_file=  'data/VOCdevkit/VOC2012/ImageSets/Main/trainval.txt',
                img_prefix= 'data/VOCdevkit/VOC2012/',
                pipeline=test_pipeline))

vocDataset2007 = build_dataset(vocData2007['train'])
vocDataset2012 = build_dataset(vocData2012['train'])
```

```python
#setup the voc2007 test dataset
vocData2007 = dict(samples_per_gpu = 4, workers_per_gpu = 4,
               val = dict(type = 'VOCDataset',
                ann_file=  'data/VOCdevkit/VOC2007/ImageSets/Main/test.txt',
                img_prefix= 'data/VOCdevkit/VOC2007/',
                pipeline=test_pipeline))

vocDataset2007Test = build_dataset(vocData2007['val'])
```

```python
#collect class per image and their filenames
imClasses = {'2007': [], '2012': [], '2007Test': []}
fileNames = {'2007': [], '2012': [], '2007Test': []}

for i in range(len(vocDataset2007)):
    imInfo = vocDataset2007.get_ann_info(i)
    clsesPresent = list(imInfo['labels']) + list(imInfo['labels_ignore'])
    imClasses['2007'] += [clsesPresent]
    fileNames['2007'] += [vocDataset2007.data_infos[i]['filename']]

for i in range(len(vocDataset2012)):
    imInfo = vocDataset2012.get_ann_info(i)
    clsesPresent = list(imInfo['labels']) + list(imInfo['labels_ignore'])
    imClasses['2012'] += [clsesPresent]
    fileNames['2012'] += [vocDataset2012.data_infos[i]['filename']]
    
for i in range(len(vocDataset2007Test)):
    imInfo = vocDataset2007Test.get_ann_info(i)
    clsesPresent = list(imInfo['labels']) + list(imInfo['labels_ignore'])
    imClasses['2007Test'] += [clsesPresent]
    fileNames['2007Test'] += [vocDataset2007Test.data_infos[i]['filename']]
```

```python
#for VOC, we take the first 15 classes as 'known' classes
cutoffCls = 14

totalNum = len(vocDataset2007) + len(vocDataset2012)
includedNum = 0
totalInstances = [0 for i in range(20)]
includedInstances = [0 for i in range(20)]
namesIncluded = {'2007': [], '2012': [], '2007Test':[]}


for yr in ['2007', '2012']:
    for idx, imCls in enumerate(imClasses[yr]):
        mask = np.asarray(imCls) > cutoffCls
        #if the image has any 'unknown' classes, it is not included in the new training dataset
        if np.sum(mask) != 0:
            for cl in imCls:
                totalInstances[cl] += 1
            continue
        #otherwise it is included in the new training dataset
        for cl in imCls:
            totalInstances[cl] += 1
            includedInstances[cl] += 1
        namesIncluded[yr] += [fileNames[yr][idx]]
        includedNum += 1

for idx, imCls in enumerate(imClasses['2007Test']):
    mask = np.asarray(imCls) > cutoffCls
    #if the image has any 'unknown' classes, it is not included in the new test dataset
    if np.sum(mask) != 0:
        continue
    #otherwise it is included in the new test dataset
    namesIncluded['2007Test'] += [fileNames['2007Test'][idx]]




```

```python
#let's check the data balance in our new training dataset
plt.figure()
plt.plot(includedInstances, label = 'New Training dataset')
plt.plot(totalInstances, label = 'Original Training dataset')
plt.ylabel('Number of instances')
plt.xlabel('Class ID')
plt.legend()
plt.title('Number of class instances between original and new training dataset')
plt.show()

#as a percent of original training dataset?
percentInstances = np.array(includedInstances)/np.array(totalInstances)
plt.figure()
plt.plot(percentInstances)
plt.ylabel('% class instances retained')
plt.xlabel('Class ID')
plt.title('Percent of class instances retained in new training dataset')
plt.show()
```

```python
#move images that don't have the unknown classes, creating our new closed-set training dataset
for yr in ['2007', '2012', '2007Test']:
    source_folder = f'data/VOCdevkit/VOC{yr}/JPEGImages/'
    destination_folder = f'data/VOCdevkit/VOC{yr}CS/JPEGImages/'

    #don't want test in the folder creation name
    if 'Test' in yr:
        destination_folder = destination_folder.replace('Test', '')
        source_folder = source_folder.replace('Test', '')
    #check destination folder exists, else create
    if not os.path.isdir(destination_folder):
        os.makedirs(destination_folder)

    for filename in tqdm(namesIncluded[yr], total = len(namesIncluded[yr])):
        nm = filename.replace('JPEGImages/', '')
        shutil.copy(os.path.join(source_folder, nm), os.path.join(destination_folder, nm))
```

```python
#Fix annotations
############################################ TRAIN/VAL ########################################################################################
for yr in ['2007', '2012']:
    for split in ['trainval', 'train', 'val']:
        print(f'Changing annotation for VOC{yr} {split} split')
        source_file = f'data/VOCdevkit/VOC{yr}/ImageSets/Main/{split}.txt'
        destination_file = f'data/VOCdevkit/VOC{yr}CS/ImageSets/Main/{split}.txt'

        if not os.path.isdir(f'data/VOCdevkit/VOC{yr}CS/ImageSets/Main'):
            os.makedirs(f'data/VOCdevkit/VOC{yr}CS/ImageSets/Main')
        
        readFile = open(source_file, 'r')
        writeFile = open(destination_file, 'w')

        for x in readFile:
            xFormat = 'JPEGImages/'+x.replace('\n', '')+'.jpg'
            if xFormat in namesIncluded[yr]:
                writeFile.write(x)
                

        writeFile.close()


print(f'Changing annotation for VOC2007 test split')
source_file = f'data/VOCdevkit/VOC2007/ImageSets/Main/test.txt'
destination_file = f'data/VOCdevkit/VOC2007CS/ImageSets/Main/test.txt'

if not os.path.isdir(f'data/VOCdevkit/VOC2007CS/ImageSets/Main'):
    os.makedirs(f'data/VOCdevkit/VOC2007CS/ImageSets/Main')

readFile = open(source_file, 'r')
writeFile = open(destination_file, 'w')

for x in readFile:
    xFormat = 'JPEGImages/'+x.replace('\n', '')+'.jpg'
    if xFormat in namesIncluded[yr]:
        writeFile.write(x)


writeFile.close()
```

# COCO

```python
COCO_CLASSES = ('person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
                'train', 'truck', 'boat', 'traffic light', 'fire hydrant',
                'stop sign', 'parking meter', 'bench', 'bird', 'cat', 'dog',
                'horse', 'sheep', 'cow', 'elephant', 'bear', 'zebra',
                'giraffe', 'backpack', 'umbrella', 'handbag', 'tie',
                'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball',
                'kite', 'baseball bat', 'baseball glove', 'skateboard',
                'surfboard', 'tennis racket', 'bottle', 'wine glass', 'cup',
                'fork', 'knife', 'spoon', 'bowl', 'banana', 'apple',
                'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
                'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed',
                'dining table', 'toilet', 'tv', 'laptop', 'mouse', 'remote',
                'keyboard', 'cell phone', 'microwave', 'oven', 'toaster', 'sink',
                'refrigerator', 'book', 'clock', 'vase', 'scissors',
                'teddy bear', 'hair drier', 'toothbrush')
```

```python
test_pipeline = [
    dict(type='LoadImageFromFile'),
    dict(
        type='MultiScaleFlipAug',
        img_scale=(608, 608),
        flip=False,
        transforms=[
            dict(type='ImageToTensor', keys=['img']),
            dict(type='Collect', keys=['img'])
        ])
]

cocoData = dict(samples_per_gpu = 8, workers_per_gpu = 4,
               train = dict(type = 'CocoDataset',
                ann_file=  'data/coco/annotations/instances_train2017.json',
                img_prefix= 'data/coco/train2017/',
                pipeline=test_pipeline),
               val = dict(type = 'CocoDataset',
                ann_file=  'data/coco/annotations/instances_val2017.json',
                img_prefix= 'data/coco/val2017/',
                pipeline=test_pipeline))

cocoTrainDataset = build_dataset(cocoData['train'])
cocoTestDataset = build_dataset(cocoData['val'])
```

```python
imClasses = {'train': [], 'test': []}
fileNames = {'train': [], 'test': []}
numIgnores = {'train': [], 'test': []}

cocoDatasets = {'train': cocoTrainDataset, 'test': cocoTestDataset}

for split in ['train', 'test']:
    for i in range(len(cocoDatasets[split])):

        imInfo = cocoDatasets[split].get_ann_info(i)
        #ignore bboxes are always crowds, which will be class person, which we will always include
        clsesPresent = list(imInfo['labels']) #+ list(imInfo['labels_ignore'])
        imClasses[split] += [clsesPresent]
        fileNames[split] += [cocoDatasets[split].data_infos[i]['filename']]
        numIgnores[split] += [len(imInfo['bboxes_ignore'])]
```

```python
#first 50 classes are known, rest are unknown
cutoffCls = 49

totalInstances = {'train':[0 for i in range(80)], 'test':[0 for i in range(80)]}
includedInstances = {'train':[0 for i in range(80)], 'test':[0 for i in range(80)]}
namesIncluded = {'train': [], 'test': []}
ignored = 0
for split in ['train', 'test']:
    for idx, imCls in enumerate(imClasses[split]):
        mask = np.asarray(imCls) > cutoffCls
        if np.sum(mask) != 0:
            for cl in imCls:
                totalInstances[split][cl] += 1

            continue
        for cl in imCls:
            totalInstances[split][cl] += 1
            includedInstances[split][cl] += 1
        namesIncluded[split] += [fileNames[split][idx]]
        ignored += numIgnores[split][idx]
```

```python
#let's check the data balance in our new training and test dataset
for split in ['train', 'test']:
    plt.figure()
    plt.plot(includedInstances[split], label = f'New {split}ing dataset')
    plt.plot(totalInstances[split], label = f'Original {split}ing dataset')
    plt.ylabel('Number of instances')
    plt.xlabel('Class ID')
    plt.legend()
    plt.title(f'Number of class instances between original and new {split}ing dataset')
    plt.show()

    #as a percent of original training dataset?
    percentInstances = np.array(includedInstances[split])/np.array(totalInstances[split])
    plt.figure()
    plt.plot(percentInstances)
    plt.ylabel('% class instances retained')
    plt.xlabel('Class ID')
    plt.title(f'Percent of class instances retained in new {split}ing dataset')
    plt.show()
```

```python
#split the training data into training data and validation data

tInstances = includedInstances['train']
totalNum = len(namesIncluded['train'])
trainNum = totalNum*0.8
namesIncluded['trainNew'] = []
namesIncluded['val'] = []
includedInstancesTrain = [0 for i in range(80)]
count = 0
for idx, imCls in enumerate(imClasses['train']):
    mask = np.asarray(imCls) > cutoffCls
    if np.sum(mask) != 0:
        continue
    if count <= trainNum:
        for cl in imCls:
           
            includedInstancesTrain[cl] += 1
        namesIncluded['trainNew'] += [fileNames['train'][idx]]
    else:
        namesIncluded['val'] += [fileNames['train'][idx]]
    count += 1
    
    
plt.figure()
plt.plot(np.array(includedInstancesTrain)/np.array(tInstances))
plt.xlabel('Class ID')
plt.ylabel('Percent in new training split - 80% desired')
plt.show()
```

```python
#move images that don't have the unknown classes present
import shutil
import os

source_folder = 'data/coco/images/train2017/'
destination_folderTrain = 'data/coco/images/trainCS2017/'
destination_folderVal = 'data/coco/images/valCS2017/'

#check destination folder exists, else create
if not os.path.isdir(destination_folderTrain):
    os.makedirs(destination_folderTrain)
    
if not os.path.isdir(destination_folderVal):
    os.makedirs(destination_folderVal)


for filename in tqdm(namesIncluded['trainNew']):
    nm = filename
    shutil.copy(os.path.join(source_folder, nm), os.path.join(destination_folderTrain, nm))
    
for filename in tqdm(namesIncluded['val']):
    nm = filename
    shutil.copy(os.path.join(source_folder, nm), os.path.join(destination_folderVal, nm))
    

source_folder = 'data/coco/images/val2017/'
destination_folderTest = 'data/coco/images/testCS2017/'
    
if not os.path.isdir(destination_folderTest):
    os.makedirs(destination_folderTest)
    
for filename in tqdm(namesIncluded['test']):
    nm = filename
    shutil.copy(os.path.join(source_folder, nm), os.path.join(destination_folderTest, nm))
```

```python
#Fix annotations
import json
source_file = '/media/dimity/7d15cfc1-e1fc-4301-9fae-a82c84c9d483/data/coco/annotations/instances_train2017.json'
destination_fileTrain = '/media/dimity/7d15cfc1-e1fc-4301-9fae-a82c84c9d483/data/coco/annotations/instances_trainCS2017.json'
destination_fileVal = '/media/dimity/7d15cfc1-e1fc-4301-9fae-a82c84c9d483/data/coco/annotations/instances_valCS2017.json'

destination_files = {'trainNew': destination_fileTrain, 'val': destination_fileVal}

with open(source_file) as f:
    readFile = json.load(f)

for split in ['trainNew', 'val']:
    print(f'Changing annotations for {split} split')
    writeFile = {}

    writeFile['info'] = readFile['info']
    writeFile['licenses'] = readFile['licenses']

    writeFile['categories'] = readFile['categories'][:50]

    writeFile['images'] = []
    writeFile['annotations'] = []

    for imKey in tqdm(readFile['images']):
        name = imKey['file_name']
        if name in namesIncluded[split]:
            writeFile['images'] += [imKey]

    for annKey in tqdm(readFile['annotations']):
        category = annKey['category_id']
        if category > 55: # this corresponds to 50th class in coco
            continue
        name = str(annKey['image_id']).zfill(12) + '.jpg'
        if name in namesIncluded[split]:
            writeFile['annotations'] += [annKey]


    with open(destination_files[split], 'w') as outFile:
        json.dump(writeFile, outFile)


#now closed-set test dataset
print(f'Changing annotations for test split')
source_file = '/media/dimity/7d15cfc1-e1fc-4301-9fae-a82c84c9d483/data/coco/annotations/instances_val2017.json'
destination_fileTest = '/media/dimity/7d15cfc1-e1fc-4301-9fae-a82c84c9d483/data/coco/annotations/instances_testCS2017.json'

with open(source_file) as f:
    readFile = json.load(f)

writeFile = {}

writeFile['info'] = readFile['info']
writeFile['licenses'] = readFile['licenses']

writeFile['categories'] = readFile['categories'][:50]

writeFile['images'] = []
writeFile['annotations'] = []


for imKey in tqdm(readFile['images']):
    name = imKey['file_name']
    if name in namesIncluded['test']:
        writeFile['images'] += [imKey]

for annKey in tqdm(readFile['annotations']):
    category = annKey['category_id']
    if category > 55: # this corresponds to 50th class in coco
        continue
    name = str(annKey['image_id']).zfill(12) + '.jpg'
    if name in namesIncluded['test']:
        writeFile['annotations'] += [annKey]


with open(destination_fileTest, 'w') as outFile:
    json.dump(writeFile, outFile)
```
